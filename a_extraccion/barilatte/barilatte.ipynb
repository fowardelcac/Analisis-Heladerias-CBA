```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
import pandas as pd

# Configurar el navegador
driver = webdriver.Chrome()
url = "https://www.google.com/maps/place/Barilatte"
driver.get(url)
```

```python
df_list = []
```

```python
# Función para extraer datos de reseñas desde una página web
def data_scraper(address: str):
    # Diccionario para almacenar los datos scrapeados
    dictionary = {
        "Usuario": [],      # Lista para nombres de usuarios
        "Rating": [],       # Lista para calificaciones
        "Fecha": [],        # Lista para fechas de las reseñas
        "Descripcion": [],  # Lista para textos de las reseñas
    }
    
    # Espera hasta que los elementos de reseñas estén presentes (máximo 10 segundos)
    WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.CSS_SELECTOR, "div.jftiEf.fontBodyMedium"))
    )
    
    # Encuentra todos los elementos de reseñas en la página
    reviews = driver.find_elements(By.CSS_SELECTOR, "div.jftiEf.fontBodyMedium")

    # Itera sobre cada reseña encontrada
    for i in reviews:
        # Extrae el nombre del usuario
        user = i.find_element(By.CSS_SELECTOR, "div.d4r55").text
        
        # Extrae la calificación (estrellas) desde el atributo aria-label
        stars = i.find_element(By.CSS_SELECTOR, "span.kvMYJc").get_attribute(
            "aria-label"
        )
        
        # Extrae la fecha de la reseña
        date = i.find_element(By.CSS_SELECTOR, "span.rsqaWe").text
        
        # Intenta extraer la descripción; si falla, asigna "None"
        try:
            description = i.find_element(By.CSS_SELECTOR, "span.wiI7pd").text
        except:
            description = "None"

        # Agrega los datos extraídos a las listas del diccionario
        dictionary["Usuario"].append(user)
        dictionary["Rating"].append(stars)
        dictionary["Fecha"].append(date)
        dictionary["Descripcion"].append(description)
        
        # Imprime confirmación de que el usuario fue añadido
        print(f"{user} añadido")

    # Convierte el diccionario en un DataFrame de pandas
    df = pd.DataFrame(dictionary)
    
    # Agrega una columna con la dirección/local proporcionada
    df["Local"] = address
    
    # Añade el DataFrame a una lista global de DataFrames
    df_list.append(df)
```
